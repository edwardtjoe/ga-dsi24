```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Q3. (Color jade OR Ideal) and carat <= 0.3, output desired: X, Y, Z
diamonds = pd.read_csv('diamonds.csv')

diamonds.loc[((diamonds["cut"] == "Ideal") | (diamonds["color"] == "J")) \
             & (diamonds["carat"] < 0.3), ["x", "y", "z"]]
```

```python
diamonds2 = diamonds.loc[((diamonds["cut"] == "Ideal") | (diamonds["color"] == "J")) \
             & (diamonds["carat"] < 0.3), :]

diamonds2
```

```python
# QUESTION: Find mean of dimensions, group by cut. (Don't use the last filter)

# Creating dimensions
diamonds["dimension"] = diamonds["x"] * diamonds["y"] * diamonds["z"]

diamonds
```

```python
diamonds.groupby("cut")["dimension"].agg(["mean", "median"])
```
