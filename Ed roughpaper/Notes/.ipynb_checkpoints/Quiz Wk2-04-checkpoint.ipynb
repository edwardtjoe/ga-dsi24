```python
def get_user_status(row):
    if row["num_rating"] == 0:
        row["class"] = "Non-rater"
    elif row["num_rating"] == 1:
        row["class"] = "Infrequent rater"
    else:
        if row["avg_rating"] > 4:
            row["class"] = "Generous rater"
        else:
            row["class"] = "Stingy rater"
    return row
```

```python
ratings.loc[
    :,
    ["UserID", "Rating"]
].groupby(
    "UserID"
).agg(
    num_rating=("Rating", "count"),
    avg_rating=("Rating", np.mean),
).reset_index(
).apply(get_user_status,
       axis=1
).merge(
    user,
    left_on="UserID",
    right_on="User_id",
    how="right"
)

```

```python
df = user.merge(ratings, left_on="User_id", right_on="UserID", how="left")

df.groupby("Username").agg(avg_rating=pd.NamedAgg(column="Rating", aggfunc=np.mean),
                           num_rating=pd.NamedAgg(column="Username", aggfunc="count")
                          ).reset_index().apply(get_user_status, axis=1)
```
