```python
#Load libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
#Load data in dataframe
train = "train.csv"
df = pd.read_csv(train)
df.head()
```

```python
#Find the average age of those who survived and those who did not
df.groupby("Survived")["Age"].mean()
```


## Step 2: Cleaning the data
1. Create a bar chart showing how many missing values are in each column
  - *Bonus* : Theres a good library for visualizing missing values called Missingno.
      - [Install Instructions](https://pypi.org/project/missingno/)
      - [Usage Documentation](https://github.com/ResidentMario/missingno)
2. Which column has the most `NaN` values? How many cells in that column are empty?
3. Delete all rows where `Embarked` is empty
4. Fill all empty cabins with **¯\\_(ツ)_/¯**

Note: `NaN`, empty, and missing are synonymous.

```python
# Q1. Create a bar chart showing how many missing values are in each column

plt.figure(figsize=(16, 5))

plt.bar(
    df.columns, 
    df.isna().sum(),
    width=1,
    label=[df.isna().sum()],
)

plt.title("Number of Missing Values");
```

```python
# This is a bar chart of how many non-missing values are present
# From this, we can visually see that the most NaN value is "Cabin"
msno.bar(df);
# Using missingno matrix function:
msno.matrix(df);
```

```python
# Q2. Which column has the most `NaN` values? How many cells in that column are empty?

f"The 'Cabin' column has the most NaN values, \
with {df['Cabin'].isnull().sum()} empty cells"
```

```python
# Q3. Delete all rows where Embarked is empty
df.dropna(subset=["Embarked"], inplace=True)
df.info()
```

```python
# Q4. Fill all empty cabins with **¯\\_(ツ)_/¯**

df["Cabin"] = df["Cabin"].fillna("¯\\_(ツ)_/¯")
df["Cabin"]
```

## Step 3: Feature extraction
1.  There are two columns that pertain to how many family members are on the boat for a given person. Create a new column called `FamilyCount` which will be the sum of those two columns.
2. Reverends have a special title in their name. Create a column called `IsReverend`: 1 if they're a preacher, 0 if they're not.
3. In order to feed our training data into a classification algorithm, we need to convert our categories into 1's and 0's using `pd.get_dummies`
  - Familiarize yourself with the [`pd.get_dummies` documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)
  - Create 3 columns: `Embarked_C`, `Embarked_Q` and `Embarked_S`. These columns will have 1's and 0's that correspond to the `C`, `Q` and `S` values in the `Embarked` column
  - Do the same thing for `Sex`
  - BONUS: Extract the title from everyone's name and create dummy columns

```python
# Q1: 1.  There are two columns that pertain to how many family members are on 
# the boat for a given person.
# Create a new column called `FamilyCount` which will be the sum of those two columns.

df["FamilyCount"] = df["SibSp"] + df["Parch"]
```

```python
# Q2: 2. Reverends have a special title in their name. Create a column called 
# `IsReverend`: 1 if they're a preacher, 0 if they're not.
df['IsReverend'] = df['Name'].str.contains('Rev.').map(lambda x: int(x))
df["IsReverend"].sum()
```

```python
# Q3: 3. In order to feed our training data into a classification algorithm, 
# we need to convert our categories into 1's and 0's using `pd.get_dummies`
# Create 3 columns: `Embarked_C`, `Embarked_Q` and `Embarked_S`. 
# These columns will have 1's and 0's that correspond to the `C`, `Q` and `S` 
# values in the `Embarked` column
# Do the same thing for `Sex`

np.unique(df["Embarked"])
```

```python
embarked_dummies = pd.get_dummies(df["Embarked"])
sex_dummies = pd.get_dummies(df["Sex"])
```

```python
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
df = pd.concat([df, sex_dummies], axis=1)
```

```python
df.head()
```

```python
# Renaming Embarked

df.rename(columns={"C": "Embarked_C", "Q": "Embarked_Q", "S": "Embarked_S"}, inplace=True)
df.head()
```

## Step 4: Exploratory analysis 
_[`df.groupby()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) may be very useful._

1. What was the survival rate overall?
2. Which gender fared the worst? What was their survival rate?
3. What was the survival rate for each `Pclass`?
4. Did any reverends survive? How many?
5. What is the survival rate for cabins marked **¯\\_(ツ)_/¯**
6. What is the survival rate for people whose `Age` is empty?
7. What is the survival rate for each port of embarkation?
8. What is the survival rate for children (under 12) in each `Pclass`?
9. Did the captain of the ship survive? Is he on the list?
10. Of all the people that died, who had the most expensive ticket? How much did it cost?
11. Does having family on the boat help or hurt your chances of survival?

```python
# 1. What was the survival rate overall?
df["Survived"].mean()

# Ans: 38.2%
```

```python
# 2. Which gender fared the worst? What was their survival rate?
df.groupby("male").mean()["Survived"]

# Ans: Males fared worst, 18.9% survival rate.
```

```python
# 3. What was the survival rate for each `Pclass`?

df.groupby("Pclass").mean()["Survived"]

# Class 1: 62.6%, Class 2: 47.3%, Class 3: 24.2%
```

```python
# 4. Did any reverends survive? How many?

# df.loc[df["IsReverend"] == 1, "Survived"].value_counts()
df.groupby("IsReverend")["Survived"].value_counts()

# Ans: No reverends survived
```

```python
# 5. What is the survival rate for cabins marked **¯\\_(ツ)_/¯**

# df.groupby("Cabin").mean()["Survived"]
df.loc[df["Cabin"] == "¯\\_(ツ)_/¯", "Survived"].mean()

# Ans: 30.0%
```

```python
# 6. What is the survival rate for people whose `Age` is empty?
df.loc[df["Age"].isnull(), "Survived"].mean()

# Ans: 29.4%
```

```python
# 7. What is the survival rate for each port of embarkation?
df.groupby("Embarked")["Survived"].mean()

# Embarked C: 55.4%
# Embarked Q: 39.0%
# Embarked S: 33.7%
```

```python
# 8. What is the survival rate for children (under 12) in each `Pclass`?

df.loc[df["Age"] < 12, "Survived"].groupby(df["Pclass"]).mean()

# Pclass 1: 75%
# Pclass 2: 100%
# Pclass 3: 40.4%
```

```python
# 9. Did the captain of the ship survive? Is he on the list?
# For loop: Captain name is Crosby, Capt. Edward Gifford

# This works:
# df.loc[df["Name"] == "Crosby, Capt. Edward Gifford", "Survived"]

# Learn how to use str.contains.
df.loc[df["Name"].str.contains("Capt.")]

# Ans: Captain of ship did not survive. He is on the list.
```

```python
# 10. Of all the people that died, who had the most expensive ticket? How much did it cost?

df.loc[df["Survived"] == 0].sort_values("Fare", ascending=False)["Name"]
# OR:
df.loc[df["Survived"] == 0].sort_values("Fare", ascending=False).head(3)

# Ans: Mr. Charles Alexander Fortune and Mr. Mark Fortune.
# Fare cost was $263 each.
```

```python
# 11. Does having family on the boat help or hurt your chances of survival?

df.groupby("FamilyCount")["Survived"].mean()

# There is an increasing rate of chance of survival when you have more 
# family members up to 3 people, after which the chances of survival drop.
```

## Step 5: Plotting
Using Matplotlib and Seaborn, create several charts showing the survival rates of different groups of people. It's fine if a handful of charts are basic (Gender, Age, etc), but what we're really looking for is something beneath the surface.


```python
df.groupby("Survived")["Fare"].mean().plot.bar()

# On average, people who survived had paid more
```

```python
survived_fare = df.loc[df["Survived"] == 1, "Fare"]
sns.displot(data=survived_fare, kde=True, bins=20);
```

```python
df.groupby("Survived")["Age"].mean().plot.bar();

# On average, people who survived were younger.
```

```python
survived_age = df.loc[df["Survived"] == 1, "Age"]
sns.displot(data=survived_age, kde=True, bins=25);
```
