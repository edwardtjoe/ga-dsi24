## Engineer a feature called `age` that shares how old the person would have been in 2014 (at the time the data was collected).

- Note: you will need to clean the data a bit.

```python
count = 0
for year in X['birth_year']:
    if pd.isna(year) is True:
        count += 1

print(count)
# Number of rows with empty 'year': 6_717
```

```python
X.dropna(subset=['birth_year'], inplace=True)
```

```python
# After: 218_019 rows
# Number of rows deleted: 6_717 Yay got this right so far
X.shape
```

```python
def age_2014(row):
    row['age'] = 2014 - row['birth_year']
    return row
```

```python
X = X.apply(age_2014, axis=1)
```

- Be sure to specify your null and alternative hypotheses, and to state your conclusion carefully and correctly **in the context of your model**!

# ANS:

Our null hypothesis is that age does not predict trip duration.

At 99% level of significance, since P < 0.01, we REJECT the null hypothesis as there is SUFFICIENT evidence that the null hypotesis is FALSE.
